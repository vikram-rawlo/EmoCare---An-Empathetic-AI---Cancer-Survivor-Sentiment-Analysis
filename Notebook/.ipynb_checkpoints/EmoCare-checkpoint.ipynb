```python
# removing punctuations from all the post

import string

def remove_punctuation(text):
    punctuationfree = "".join([i for i in text if i not in string.punctuation])
    return punctuationfree
```

```python
# Removing stopwords
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))

def remove_stopwords(text):
    if isinstance(text, list):
        text = ' '.join(text)
    output = " ".join(word for word in text.split() if word.lower() not in stop_words)
    return output
```

# Text Cleaning

```python
from tqdm import tqdm
import contractions
from bs4 import BeautifulSoup
from nltk.stem import WordNetLemmatizer
import re
import nltk

wordnet = WordNetLemmatizer()

stopwords = nltk.corpus.stopwords.words('english')

preprocessed_reviews = []

for sentence in tqdm(df1['posts'].values):
  sentence = contractions.fix(sentence)
  sentence = BeautifulSoup(sentence, 'lxml').get_text()
  sentence = re.sub(r"https\S+","", sentence)
  sentence = re.sub("\S*\d\S*","", sentence).strip()
  sentence = re.sub('[^a-zA-Z\s]+'," ", sentence)
  tokens = sentence.split()
  tokens = [wordnet.lemmatize(word) for word in tokens if word.lower() not in stopwords]
  cleaned_sentence = ' '.join(tokens).lower()
  preprocessed_reviews.append(cleaned_sentence.strip())
```

## Model building using Deep learning

```python
import tensorflow as tf
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense, Dropout
from tensorflow.keras.regularizers import l2
```

```python
# Splitting training & validation set using stratify
x_train_resampled, x_val_resampled, y_train_resampled, y_val_resampled = train_test_split(x_train, y_train, test_size=0.2, stratify=y_train)
```

```python
input_dim = x_train_resampled.shape[1]

mlp_model = Sequential()
mlp_model.add(Dense(256, input_dim=input_dim, activation='relu', kernel_regularizer=l2(0.01)))
mlp_model.add(Dropout(0.20))
mlp_model.add(Dense(128, activation='relu', kernel_regularizer=l2(0.01)))
mlp_model.add(Dropout(0.20))
mlp_model.add(Dense(64, activation='relu', kernel_regularizer=l2(0.01)))
mlp_model.add(Dropout(0.20))
mlp_model.add(Dense(4, activation='softmax'))
```

```python
mlp_model.compile(loss='sparse_categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

```python
mlp_model.fit(x_train_resampled, y_train_resampled, epochs=50, batch_size=32, validation_data=(x_val_resampled,y_val_resampled))
```

```python
y_pred_train_mlp = mlp_model.predict(x_train_resampled)
y_pred_train_classes = y_pred_train_mlp.argmax(axis=1)
y_pred_test_mlp = mlp_model.predict(x_test)
y_pred_test_classes = y_pred_test_mlp.argmax(axis=1)
```

```python
model_evaluation(y_train_resampled, y_pred_train_classes, y_test, y_pred_test_classes)
```
